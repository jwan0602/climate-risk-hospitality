```python
import pandas as pd

# 🔹 1. 파일 불러오기
cbre = pd.read_csv("C:/Users/jwan0/Downloads/Climate risk/CBRE 2007-2023.csv")
local = pd.read_csv("C:/Users/jwan0/Downloads/Climate risk/Climate_Risk_Index_Local.csv")
regional = pd.read_csv("C:/Users/jwan0/Downloads/Climate risk/Climate_Risk_Index_Regional.csv")

# 🔹 2. ZIP, STATE, YEAR 전처리
cbre['ZIP'] = cbre['Zip'].astype(str).str.replace('.0', '', regex=False).str.zfill(5)
cbre['YEAR'] = cbre['Year of Data'].astype(int)
cbre['STATE'] = cbre['State'].str.upper()

local['ZIP'] = local['ZIP'].astype(str).str.zfill(5)
local['YEAR'] = local['YEAR'].astype(int)

regional['year'] = regional['year'].astype(int)
regional['STATE'] = regional['STATE_ABBR'].str.upper()

# 🔹 3. ZIP 기준 국지 지수 병합
merged_local = cbre.merge(local[['ZIP', 'YEAR', 'Climate_Risk_Index_Local']],
                          on=['ZIP', 'YEAR'], how='left')

# 🔹 4. STATE 기준 광역 지수 병합
merged_final = merged_local.merge(regional[['STATE', 'year', 'Climate_Risk_Index_Regional']],
                                  left_on=['STATE', 'YEAR'], right_on=['STATE', 'year'], how='left')

# 🔹 5. 중복 열 제거 및 정리
merged_final.drop(columns=['year'], inplace=True)

# 🔹 6. 결과 미리보기 및 저장
print(merged_final[['PropertyID', 'ZIP', 'STATE', 'YEAR',
                    'Climate_Risk_Index_Local', 'Climate_Risk_Index_Regional']].head())

# 🔹 7. 결과 저장
merged_final.to_csv("CBRE_with_Climate_Risk_Index.csv", index=False)
```
